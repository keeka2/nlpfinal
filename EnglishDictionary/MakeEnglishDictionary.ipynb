```python
#의미있는 영어만 추출하여 영어 사전 만들기
import csv
import pandas as pd
import re

file = open('hashcode_classification2020_train.csv', 'r', encoding='utf-8')
line = csv.reader(file)

f1 = open("EnglishCount_C_1.txt", 'w')
f2 = open("EnglishCount_C++_2.txt", 'w')
f3 = open("EnglishCount_JAVA_3.txt", 'w')
f4 = open("EnglishCount_JAVASCRIPT_4.txt", 'w')
f5 = open("EnglishCount_PYTHON_5.txt", 'w')

    
for line_text in line:
    if line_text[2]=='1':
        text=re.sub('[^a-zA-Z]',' ',line_text[0]+line_text[1]).strip()
        temp=text.split(' ')
        temp = list(filter(None, temp))
        s=' '.join(temp)
        f1.write(s)
        f1.write('\n')
    elif line_text[2]=='2':
        text=re.sub('[^a-zA-Z]',' ',line_text[0]+line_text[1]).strip()
        temp=text.split(' ')
        temp = list(filter(None, temp))
        s=' '.join(temp)
        f2.write(s)
        f2.write('\n')
    elif line_text[2]=='3':
        text=re.sub('[^a-zA-Z]',' ',line_text[0]+line_text[1]).strip()
        temp=text.split(' ')
        temp = list(filter(None, temp))
        s=' '.join(temp)
        f3.write(s)
        f3.write('\n')
    elif line_text[2]=='4':
        text=re.sub('[^a-zA-Z]',' ',line_text[0]+line_text[1]).strip()
        temp=text.split(' ')
        temp = list(filter(None, temp))
        s=' '.join(temp)
        f4.write(s)
        f4.write('\n')
    elif line_text[2]=='5':
        text=re.sub('[^a-zA-Z]',' ',line_text[0]+line_text[1]).strip()
        temp=text.split(' ')
        temp = list(filter(None, temp))
        s=' '.join(temp)
        f5.write(s)
        f5.write('\n')
    
file.close()
f1.close()
f2.close()
f3.close()
f4.close()
f5.close()
```

```python
from collections import Counter
wordDict1 = Counter()
wordDict2 = Counter()
wordDict3 = Counter()
wordDict4 = Counter()
wordDict5 = Counter()

f1 = open("EnglishCount_C_1.txt", 'r')
f2 = open("EnglishCount_C++_2.txt", 'r')
f3 = open("EnglishCount_JAVA_3.txt", 'r')
f4 = open("EnglishCount_JAVASCRIPT_4.txt", 'r')
f5 = open("EnglishCount_PYTHON_5.txt", 'r')

#총 단어 수
count1=0
count2=0
count3=0
count4=0
count5=0

sentences = f1.readlines()
for sentence in sentences: #한 문장씩
    for word in sentence.split(): #한 문장에 들어있는 한 단어씩
        count1+=1
        wordDict1[word] += 1 #Counter에 count를 1씩 증가시킨다.
        
sentences = f2.readlines()
for sentence in sentences: #한 문장씩
    for word in sentence.split(): #한 문장에 들어있는 한 단어씩
        count2+=1
        wordDict2[word] += 1 #Counter에 count를 1씩 증가시킨다.
        
sentences = f3.readlines()
for sentence in sentences: #한 문장씩
    for word in sentence.split(): #한 문장에 들어있는 한 단어씩
        count3+=1
        wordDict3[word] += 1 #Counter에 count를 1씩 증가시킨다.
        
sentences = f4.readlines()
for sentence in sentences: #한 문장씩
    for word in sentence.split(): #한 문장에 들어있는 한 단어씩
        count4+=1
        wordDict4[word] += 1 #Counter에 count를 1씩 증가시킨다.
        
sentences = f5.readlines()
for sentence in sentences: #한 문장씩
    for word in sentence.split(): #한 문장에 들어있는 한 단어씩
        count5+=1
        wordDict5[word] += 1 #Counter에 count를 1씩 증가시킨다.

        
#각 언어별 단어 빈도수 세기  
f1_1 = open("EnglishCountWord_C_1.txt", 'w')
f2_2 = open("EnglishCountWord_C++_2.txt", 'w')
f3_3 = open("EnglishCountWord_JAVA_3.txt", 'w')
f4_4 = open("EnglishCountWord_JAVASCRIPT_4.txt", 'w')
f5_5 = open("EnglishCountWord_PYTHON_5.txt", 'w')


for word, freq in wordDict1.most_common(100000):
    if freq>4:
        f1_1.write(word+' '+str((freq/count1)*100))
        f1_1.write('\n')
for word, freq in wordDict2.most_common(100000):
    if freq>5:
        f2_2.write(word+' '+str((freq/count2)*100))
        f2_2.write('\n')
for word, freq in wordDict3.most_common(100000):
    if freq>19:
        f3_3.write(word+' '+str((freq/count3)*100))
        f3_3.write('\n')
for word, freq in wordDict4.most_common(100000):
    if freq>8:
        f4_4.write(word+' '+str((freq/count4)*100))
        f4_4.write('\n')
for word, freq in wordDict5.most_common(100000):
    if freq>11:
        f5_5.write(word+' '+str((freq/count5)*100))
        f5_5.write('\n')
        
f1.close()
f2.close()
f3.close()
f4.close()
f5.close()
f1_1.close()
f2_2.close()
f3_3.close()
f4_4.close()
f5_5.close()
```
