```python
#의미있는 특수문자만 추출
import csv
import pandas as pd
import re

file = open('hashcode_classification2020_train.csv', 'r', encoding='utf-8')
line = csv.reader(file)

f1 = open("SpecialCount_C_1.txt", 'w', encoding='utf-8')
f2 = open("SpecialCount_C++_2.txt", 'w', encoding='utf-8')
f3 = open("SpecialCount_JAVA_3.txt", 'w', encoding='utf-8')
f4 = open("SpecialCount_JAVASCRIPT_4.txt", 'w', encoding='utf-8')
f5 = open("SpecialCount_PYTHON_5.txt", 'w', encoding='utf-8')

count=0
for line_text in line:
    text=re.sub('[a-zA-Z0-9]',' ',line_text[0]+line_text[1])
    text=re.sub('[ㄱ-ㅣ가-힣]+',' ',text)
    if line_text[2]=='1':
        temp=text.split(' ')
        temp = list(filter(None, temp))
        s=' '.join(temp)
        f1.write(s)
        f1.write('\n')
    elif line_text[2]=='2':
        temp=text.split(' ')
        temp = list(filter(None, temp))
        s=' '.join(temp)
        f2.write(s)
        f2.write('\n')
    elif line_text[2]=='3':
        temp=text.split(' ')
        temp = list(filter(None, temp))
        s=' '.join(temp)
        f3.write(s)
        f3.write('\n')
    elif line_text[2]=='4':
        temp=text.split(' ')
        temp = list(filter(None, temp))
        s=' '.join(temp)
        f4.write(s)
        f4.write('\n')
    elif line_text[2]=='5':
        temp=text.split(' ')
        temp = list(filter(None, temp))
        s=' '.join(temp)
        f5.write(s)
        f5.write('\n')
file.close()
f1.close()
f2.close()
f3.close()
f4.close()
f5.close()
```

```python
from collections import Counter
wordDict1 = Counter()
wordDict2 = Counter()
wordDict3 = Counter()
wordDict4 = Counter()
wordDict5 = Counter()

f1 = open("SpecialCount_C_1.txt", 'r', encoding='utf-8')
f2 = open("SpecialCount_C++_2.txt", 'r', encoding='utf-8')
f3 = open("SpecialCount_JAVA_3.txt", 'r', encoding='utf-8')
f4 = open("SpecialCount_JAVASCRIPT_4.txt", 'r', encoding='utf-8')
f5 = open("SpecialCount_PYTHON_5.txt", 'r', encoding='utf-8')

#총 단어 수
count1=0
count2=0
count3=0
count4=0
count5=0

sentences = f1.readlines()
for sentence in sentences: #한 문장씩
    for word in sentence.split(): #한 문장에 들어있는 한 단어씩
        if '	' not in word:
            count1+=1
            wordDict1[word] += 1 #Counter에 count를 1씩 증가시킨다.
        
sentences = f2.readlines()
for sentence in sentences: #한 문장씩
    for word in sentence.split(): #한 문장에 들어있는 한 단어씩
        if '	' not in word:
            count2+=1
            wordDict2[word] += 1 #Counter에 count를 1씩 증가시킨다.
        
sentences = f3.readlines()
for sentence in sentences: #한 문장씩
    for word in sentence.split(): #한 문장에 들어있는 한 단어씩
        if '	' not in word:
            count3+=1
            wordDict3[word] += 1 #Counter에 count를 1씩 증가시킨다.
        
sentences = f4.readlines()
for sentence in sentences: #한 문장씩
    for word in sentence.split(): #한 문장에 들어있는 한 단어씩
        if '	' not in word:
            count4+=1
            wordDict4[word] += 1 #Counter에 count를 1씩 증가시킨다.
        
sentences = f5.readlines()
for sentence in sentences: #한 문장씩
    for word in sentence.split(): #한 문장에 들어있는 한 단어씩
        if '	' not in word:
            count5+=1
            wordDict5[word] += 1 #Counter에 count를 1씩 증가시킨다.

        
#각 언어별 단어 빈도수 세기  
f1_1 = open("SpecialCountWord_C_1.txt", 'w', encoding='utf-8')
f2_2 = open("SpecialCountWord_C++_2.txt", 'w', encoding='utf-8')
f3_3 = open("SpecialCountWord_JAVA_3.txt", 'w', encoding='utf-8')
f4_4 = open("SpecialCountWord_JAVASCRIPT_4.txt", 'w', encoding='utf-8')
f5_5 = open("SpecialCountWord_PYTHON_5.txt", 'w', encoding='utf-8')


for word, freq in wordDict1.most_common(100000):
    if freq>10:
        f1_1.write(word+' '+str((freq/count1)*100))
        f1_1.write('\n')
for word, freq in wordDict2.most_common(100000):
    if freq>9:
        f2_2.write(word+' '+str((freq/count2)*100))
        f2_2.write('\n')
for word, freq in wordDict3.most_common(100000):
    if freq>25:
        f3_3.write(word+' '+str((freq/count3)*100))
        f3_3.write('\n')
for word, freq in wordDict4.most_common(100000):
    if freq>13:
        f4_4.write(word+' '+str((freq/count4)*100))
        f4_4.write('\n')
for word, freq in wordDict5.most_common(100000):
    if freq>19:
        f5_5.write(word+' '+str((freq/count5)*100))
        f5_5.write('\n')
        
f1.close()
f2.close()
f3.close()
f4.close()
f5.close()
f1_1.close()
f2_2.close()
f3_3.close()
f4_4.close()
f5_5.close()
```

```python
#큰 의미 없는 특수문자 제거(빈도수 차이가 다른 것과 비교하여 얼마 안나면 제거)  
f1_1 = open("SpecialCountWord_C_1.txt", 'r', encoding='utf-8')
f2_2 = open("SpecialCountWord_C++_2.txt", 'r', encoding='utf-8')
f3_3 = open("SpecialCountWord_JAVA_3.txt", 'r', encoding='utf-8')
f4_4 = open("SpecialCountWord_JAVASCRIPT_4.txt", 'r', encoding='utf-8')
f5_5 = open("SpecialCountWord_PYTHON_5.txt", 'r', encoding='utf-8')


dict_C={}
dict_CPP={}
dict_JAVA={}
dict_JAVASCRIPT={}
dict_PYTHON={}

sentences = f1_1.readlines()
for sentence in sentences: #한 문장씩
    temp = sentence.split() #한 문장에 들어있는 한 단어씩
    dict_C[temp[0]]=temp[1]

sentences = f2_2.readlines()
for sentence in sentences: #한 문장씩
    temp = sentence.split() #한 문장에 들어있는 한 단어씩
    dict_CPP[temp[0]]=temp[1]

sentences = f3_3.readlines()
for sentence in sentences: #한 문장씩
    temp = sentence.split() #한 문장에 들어있는 한 단어씩
    dict_JAVA[temp[0]]=temp[1]
    
sentences = f4_4.readlines()
for sentence in sentences: #한 문장씩
    temp = sentence.split() #한 문장에 들어있는 한 단어씩
    dict_JAVASCRIPT[temp[0]]=temp[1]
    
sentences = f5_5.readlines()
for sentence in sentences: #한 문장씩
    temp = sentence.split() #한 문장에 들어있는 한 단어씩
    dict_PYTHON[temp[0]]=temp[1]


f_1 = open("Special_Preprocess_C_1.txt", 'w', encoding='utf-8')
f_2 = open("Special_Preprocess_C++_2.txt", 'w', encoding='utf-8')
f_3 = open("Special_Preprocess_JAVA_3.txt", 'w', encoding='utf-8')
f_4 = open("Special_Preprocess_JAVASCRIPT_4.txt", 'w', encoding='utf-8')
f_5 = open("Special_Preprocess_PYTHON_5.txt", 'w', encoding='utf-8')


#C언어 주요 단어 분류(특수문자)
final_c=[]
for i in dict_C:
    #print(i+' '+dict_C[i])
    #같은 단어
    c=0
    if i in dict_CPP:
        #비율 비교
        if float(dict_C[i])<float(dict_CPP[i]):
            difference=float(dict_C[i])/float(dict_CPP[i])
        else:
            difference=float(dict_CPP[i])/float(dict_C[i])

        if difference > 0.3:
            c+=1
            
    if i in dict_JAVA:
        if float(dict_C[i])<float(dict_JAVA[i]):
            difference=float(dict_C[i])/float(dict_JAVA[i])
        else:
            difference=float(dict_JAVA[i])/float(dict_C[i])

        if difference > 0.3:
            c+=1
            
    if i in dict_JAVASCRIPT:
        if float(dict_C[i])<float(dict_JAVASCRIPT[i]):
            difference=float(dict_C[i])/float(dict_JAVASCRIPT[i])
        else:
            difference=float(dict_JAVASCRIPT[i])/float(dict_C[i])

        if difference > 0.3:
            c+=1
            
    if i in dict_PYTHON:
        if float(dict_C[i])<float(dict_PYTHON[i]):
            difference=float(dict_C[i])/float(dict_PYTHON[i])
        else:
            difference=float(dict_PYTHON[i])/float(dict_C[i])

        if difference > 0.3:
            c+=1
    if c<2:
        final_c.append(i)
for final in final_c:
    f_1.write(final)
    f_1.write('\n')
    
    
#C++언어 주요 단어 분류(특수문자)
final_cpp=[]
for i in dict_CPP:
    #print(i+' '+dict_C[i])
    #같은 단어
    c=0
    if i in dict_C:
        #비율 비교
        if float(dict_CPP[i])<float(dict_C[i]):
            difference=float(dict_CPP[i])/float(dict_C[i])
        else:
            difference=float(dict_C[i])/float(dict_CPP[i])

        if difference > 0.3:
            c+=1
            
    if i in dict_JAVA:
        if float(dict_CPP[i])<float(dict_JAVA[i]):
            difference=float(dict_CPP[i])/float(dict_JAVA[i])
        else:
            difference=float(dict_JAVA[i])/float(dict_CPP[i])

        if difference > 0.3:
            c+=1
            
    if i in dict_JAVASCRIPT:
        if float(dict_CPP[i])<float(dict_JAVASCRIPT[i]):
            difference=float(dict_CPP[i])/float(dict_JAVASCRIPT[i])
        else:
            difference=float(dict_JAVASCRIPT[i])/float(dict_CPP[i])

        if difference > 0.3:
            c+=1
            
    if i in dict_PYTHON:
        if float(dict_CPP[i])<float(dict_PYTHON[i]):
            difference=float(dict_CPP[i])/float(dict_PYTHON[i])
        else:
            difference=float(dict_PYTHON[i])/float(dict_CPP[i])

        if difference > 0.3:
            c+=1
    if c<2:
        final_cpp.append(i)
for final in final_cpp:
    f_2.write(final)
    f_2.write('\n')
    
#JAVA언어 주요 단어 분류(특수문자)
final_java=[]
for i in dict_JAVA:
    #print(i+' '+dict_C[i])
    #같은 단어
    c=0
    if i in dict_CPP:
        #비율 비교
        if float(dict_JAVA[i])<float(dict_CPP[i]):
            difference=float(dict_JAVA[i])/float(dict_CPP[i])
        else:
            difference=float(dict_CPP[i])/float(dict_JAVA[i])

        if difference > 0.3:
            c+=1
            
    if i in dict_C:
        if float(dict_JAVA[i])<float(dict_C[i]):
            difference=float(dict_JAVA[i])/float(dict_C[i])
        else:
            difference=float(dict_C[i])/float(dict_JAVA[i])

        if difference > 0.3:
            c+=1
            
    if i in dict_JAVASCRIPT:
        if float(dict_JAVA[i])<float(dict_JAVASCRIPT[i]):
            difference=float(dict_JAVA[i])/float(dict_JAVASCRIPT[i])
        else:
            difference=float(dict_JAVASCRIPT[i])/float(dict_JAVA[i])

        if difference > 0.3:
            c+=1
            
    if i in dict_PYTHON:
        if float(dict_JAVA[i])<float(dict_PYTHON[i]):
            difference=float(dict_JAVA[i])/float(dict_PYTHON[i])
        else:
            difference=float(dict_PYTHON[i])/float(dict_JAVA[i])

        if difference > 0.3:
            c+=1
    if c<2:
        final_java.append(i)
for final in final_java:
    f_3.write(final)
    f_3.write('\n')
    
#JAVASCRIPT언어 주요 단어 분류(특수문자)
final_javascript=[]
for i in dict_JAVASCRIPT:
    #print(i+' '+dict_C[i])
    #같은 단어
    c=0
    if i in dict_CPP:
        #비율 비교
        if float(dict_JAVASCRIPT[i])<float(dict_CPP[i]):
            difference=float(dict_JAVASCRIPT[i])/float(dict_CPP[i])
        else:
            difference=float(dict_CPP[i])/float(dict_JAVASCRIPT[i])

        if difference > 0.3:
            c+=1
            
    if i in dict_JAVA:
        if float(dict_JAVASCRIPT[i])<float(dict_JAVA[i]):
            difference=float(dict_JAVASCRIPT[i])/float(dict_JAVA[i])
        else:
            difference=float(dict_JAVA[i])/float(dict_JAVASCRIPT[i])

        if difference > 0.3:
            c+=1
            
    if i in dict_C:
        if float(dict_JAVASCRIPT[i])<float(dict_C[i]):
            difference=float(dict_JAVASCRIPT[i])/float(dict_C[i])
        else:
            difference=float(dict_C[i])/float(dict_JAVASCRIPT[i])

        if difference > 0.3:
            c+=1
            
    if i in dict_PYTHON:
        if float(dict_JAVASCRIPT[i])<float(dict_PYTHON[i]):
            difference=float(dict_JAVASCRIPT[i])/float(dict_PYTHON[i])
        else:
            difference=float(dict_PYTHON[i])/float(dict_JAVASCRIPT[i])

        if difference > 0.3:
            c+=1
    if c<2:
        final_javascript.append(i)
for final in final_javascript:
    f_4.write(final)
    f_4.write('\n')
    
#PYTHON언어 주요 단어 분류(특수문자)
final_python=[]
for i in dict_PYTHON:
    #print(i+' '+dict_C[i])
    #같은 단어
    c=0
    if i in dict_CPP:
        #비율 비교
        if float(dict_PYTHON[i])<float(dict_CPP[i]):
            difference=float(dict_PYTHON[i])/float(dict_CPP[i])
        else:
            difference=float(dict_CPP[i])/float(dict_PYTHON[i])

        if difference > 0.3:
            c+=1
            
    if i in dict_JAVA:
        if float(dict_PYTHON[i])<float(dict_JAVA[i]):
            difference=float(dict_PYTHON[i])/float(dict_JAVA[i])
        else:
            difference=float(dict_JAVA[i])/float(dict_PYTHON[i])

        if difference > 0.3:
            c+=1
            
    if i in dict_JAVASCRIPT:
        if float(dict_PYTHON[i])<float(dict_JAVASCRIPT[i]):
            difference=float(dict_PYTHON[i])/float(dict_JAVASCRIPT[i])
        else:
            difference=float(dict_JAVASCRIPT[i])/float(dict_PYTHON[i])

        if difference > 0.3:
            c+=1
            
    if i in dict_C:
        if float(dict_PYTHON[i])<float(dict_C[i]):
            difference=float(dict_PYTHON[i])/float(dict_C[i])
        else:
            difference=float(dict_C[i])/float(dict_PYTHON[i])

        if difference > 0.3:
            c+=1
    if c<2:
        final_python.append(i)
for final in final_python:
    f_5.write(final)
    f_5.write('\n')
    


f1_1.close()
f2_2.close()
f3_3.close()
f4_4.close()
f5_5.close()
f_1.close()
f_2.close()
f_3.close()
f_4.close()
f_5.close()
```
